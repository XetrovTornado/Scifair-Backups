# CSE-CIC-IDS 2017 Exploratory Data Analysis with PCA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

import glob
```

```python
NOTEBOOK_PATH = "C:/Users/Xetrov/Desktop/SciFair20/Code/"
```

```python
x_scaled = pd.read_csv(NOTEBOOK_PATH + "IDS2017/x_scaled_powertransform.csv")
```

```python
y_df_1h = pd.read_csv(NOTEBOOK_PATH + "IDS2017/y_grouped_1henc.csv")
```

```python
y_df = np.argmax(y_df_1h.to_numpy(), axis=1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(x_scaled)
```

```python
x_pca = pca.transform(x_scaled)
```

```python
x_scaled.shape
```

```python
x_pca.shape
```

```python
np.unique(y_df)
```

```python
pca.explained_variance_
```

```python
%%time
plt.figure(figsize=(16,12))
plt.scatter(x_pca[:,0], x_pca[:,1], c=y_df, label=y_df, cmap='plasma', s=0.1, alpha=0.2)
plt.legend()
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
