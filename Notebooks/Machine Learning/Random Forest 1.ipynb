# CSE-CIC-IDS 2017 Random Forests

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

import glob
```

```python
NOTEBOOK_PATH = "D:/Delta Stuff/Scifair20/"
```

```python
f_names = glob.glob(NOTEBOOK_PATH + "IDS2017/MachineLearningCVE/*.csv")
df = pd.concat(tqdm(pd.read_csv(f) for f in f_names))
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.columns = list(map(lambda x: x.strip(), df.columns))
```

```python
# Remove variables with no variance
remove_vars = []
for i in tqdm(range(len(df.columns))):
    col = df.columns[i]
    unique = df[col].unique()
    if len(unique) <= 1:
        remove_vars.append(col)
```

```python
remove_vars
```

```python
df.drop(remove_vars, axis=1, inplace=True)
```

```python
df.shape
```

```python
df.drop(["Flow Bytes/s", "Flow Packets/s"], axis='columns', inplace=True)
```

```python
df.shape
```

# Standardize data

```python
from sklearn.preprocessing import PowerTransformer
```

```python
scaler = PowerTransformer(method='yeo-johnson', standardize=True)
```

```python
x_df = df.drop("Label", axis=1)
y_df = df.loc[:, "Label"]
```

```python
x_df_ptscaled = scaler.fit_transform(x_df, y_df)
```

```python
scaled_summary = pd.DataFrame(x_df_ptscaled, columns=x_df.columns).describe()
scaled_summary
```

```python
y_df_group = y_df.copy()

for i in tqdm(range(len(y_df_group))):
    attack = y_df_group[i]
    group = attack
    if attack in ("DoS Hulk", "DoS GoldenEye", "DoS slowloris", "DoS Slowhttptest"):
        group = "DoS"
    elif attack.startswith("Web Attack"):
        group = "Web Attack"
    elif attack == "FTP-Patator":
        group = "Brute Force FTP"
    elif attack == "SSH-Patator":
        group = "Brute Force SSH"
    
    y_df_group[i] = group
```

```python
y_df.value_counts()
```

```python
y_df_group.value_counts()
```

```python
y_df_enc = pd.get_dummies(y_df_group)
```

```python
y_df_enc
```

# Split data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_df_ptscaled, y_df_enc, test_size = 0.1, random_state = 42)
```

```python
y_train.sum(axis=0)
```

```python
y_val.sum(axis=0)
```

# Train model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42, verbose=2).fit(x_train, y_train)
```

```python
model_id = "randomforest1"
```

**Save model**

```python
import pickle
```

```python
import time

model_filename = model_id  + " [" + time.strftime("%Y%m%d %H%M") + "]"
```

```python
save_file = open(NOTEBOOK_PATH + "Models/" + model_filename + ".pkl", "wb")
saved_model = pickle.dump(clf, save_file)
save_file.close()
```

**Test model**

```python
pred = clf.predict(x_val)
```

```python
pred.sum(axis=0)
```

```python
y_val.sum(axis=0)
```

```python
pred_index = list(pred.argmax(1))
y_index = list(np.argmax(y_val.to_numpy(), axis=1))

atktypes = sorted(y_val.columns)
pred_series = pd.Series(pred_index, name="Pred").replace(dict(enumerate(atktypes)))
y_series = pd.Series(y_index, name="Actual").replace(dict(enumerate(atktypes)))

matrix = pd.crosstab(pred_series, y_series, margins=True)
matrix
```

# F1 Score

```python
from sklearn.metrics import f1_score
```

```python
f1_micro = f1_score(y_index, pred_index, average='micro')
f1_macro = f1_score(y_index, pred_index, average='macro')

print(f1_micro)
print(f1_macro)
```

# Log results

```python
model_log = open(NOTEBOOK_PATH + "model_log.txt", "a")

model_log.write("\n" + model_filename)
model_log.write("\n\tF1 Micro: " + str(f1_micro))
model_log.write("\n\tF1 Macro: " + str(f1_macro))

model_log.close()
```
